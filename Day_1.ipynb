```python
import numpy as np
import pandas as pd
import seaborn as sns

```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days = [1,2,3,4,5,6,7]
sales_one = [160, 150, 140, 145, 175, 165, 180]
sales_two = [70, 90, 160, 150, 140, 145, 175]
plt.figure(figsize=(1,2), dpi=100)
plt.plot(days, sales_one)
plt.plot(days, sales_two)
plt.title('Assignment 2')
plt.xlabel('days of the week')
plt.ylabel('sales')
plt.grid()
plt.show()
```
