```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline

from numpy.random import randint, randn , uniform , sample
```

```python
#Relplot

fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
sns.relplot(x="timepoint", y='signal',data= fmri)
```

```python
sns.relplot(x="timepoint", y='signal',data= fmri, kind ="line", hue ="region" )
```

```python
sns.relplot(x="timepoint", y='signal',data= fmri, kind ="line",style= "event" )
```
